```python
import os
import cv2

input_path = r"D:\Complete_Data\research\PCOS"
output_path_raw = r"D:\Complete_Data\research\PCOS_processed_image\raw"
os.makedirs(output_path_raw, exist_ok=True)

target_size = (224, 224)  # or 256x256 based on your model

def preprocess_raw_images(input_dir, output_dir):
    for label in ['infected', 'noninfected']:  # assuming folder names
        folder_in = os.path.join(input_dir, label)
        folder_out = os.path.join(output_dir, label)
        os.makedirs(folder_out, exist_ok=True)

        for file_name in os.listdir(folder_in):
            img_path = os.path.join(folder_in, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # ultrasound often grayscale
            if img is None:
                continue
            img_resized = cv2.resize(img, target_size)
            # Normalize to 0-1 float
            img_normalized = img_resized / 255.0
            # Save normalized image as PNG or JPG (scaled back to 0-255 for saving)
            save_path = os.path.join(folder_out, file_name)
            cv2.imwrite(save_path, (img_normalized * 255).astype('uint8'))

preprocess_raw_images(input_path, output_path_raw)
```

```python
import os
import shutil
import random

# Set split sizes based on smallest class (noninfected: 5000 images)
train_size = 3750
val_size = 1000
test_size = 250

def split_data_fixed(source_dir, train_dir, val_dir, test_dir, train_size, val_size, test_size):
    files = [f for f in os.listdir(source_dir) if os.path.isfile(os.path.join(source_dir, f))]
    random.shuffle(files)
    total_needed = train_size + val_size + test_size
    if len(files) < total_needed:
        raise ValueError(f"Not enough files in {source_dir}: required {total_needed}, found {len(files)}")
    train_files = files[:train_size]
    val_files = files[train_size:train_size+val_size]
    test_files = files[train_size+val_size:train_size+val_size+test_size]
    for path in [train_dir, val_dir, test_dir]:
        os.makedirs(path, exist_ok=True)
    for f in train_files:
        shutil.copy(os.path.join(source_dir, f), os.path.join(train_dir, f))
    for f in val_files:
        shutil.copy(os.path.join(source_dir, f), os.path.join(val_dir, f))
    for f in test_files:
        shutil.copy(os.path.join(source_dir, f), os.path.join(test_dir, f))

# Define your source and destination folders
infected_source = r"D:/Complete_Data/research/PCOS_processed_image/raw/infected"
noninfected_source = r"D:/Complete_Data/research/PCOS_processed_image/raw/noninfected"
base_split_dir = r"D:/Complete_Data/research/PCOS_processed_image/split_balanced_image"

# Split infected
split_data_fixed(
    infected_source,
    os.path.join(base_split_dir, "train/infected"),
    os.path.join(base_split_dir, "val/infected"),
    os.path.join(base_split_dir, "test/infected"),
    train_size, val_size, test_size
)

# Split noninfected
split_data_fixed(
    noninfected_source,
    os.path.join(base_split_dir, "train/noninfected"),
    os.path.join(base_split_dir, "val/noninfected"),
    os.path.join(base_split_dir, "test/noninfected"),
    train_size, val_size, test_size
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import os

# Update the path to your new split_balanced_image directory
base_split_dir = r"D:\Complete_Data\research\PCOS_processed_image\split_balanced_image"
train_minority_dir = os.path.join(base_split_dir, "train/noninfected")
target_train_count = 3750  # Set this to your desired balanced train size

files = [f for f in os.listdir(train_minority_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))]
orig_count = len(files)
to_generate = target_train_count - orig_count

if to_generate > 0:
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.1,
        zoom_range=0.1,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    gen_count = 0
    while gen_count < to_generate:
        for fname in files:
            img_path = os.path.join(train_minority_dir, fname)
            img = load_img(img_path)
            x = img_to_array(img)
            x = x.reshape((1,) + x.shape)
            for batch in datagen.flow(x, batch_size=1, save_to_dir=train_minority_dir,
                                      save_prefix='aug', save_format='png'):
                gen_count += 1
                if gen_count >= to_generate:
                    break
            if gen_count >= to_generate:
                break
    print(f"Augmentation completed. Total images: {orig_count + gen_count}")
else:
    print("No augmentation needed, dataset already balanced in training set.")
```

```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, CSVLogger

# Set data directory and parameters
data_dir = r"D:\Complete_Data\research\PCOS_processed_image\split_balanced_image"
img_size = (224, 224)
batch_size = 16  # Use a smaller batch size for CPU efficiency
epochs = 50

# Data generators with light augmentation for training only
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)
val_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    os.path.join(data_dir, "train"),
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)
val_gen = val_datagen.flow_from_directory(
    os.path.join(data_dir, "val"),
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)

# Build ResNet50 model with dropout for regularization
base_resnet = ResNet50(weights='imagenet', include_top=False, input_shape=img_size + (3,))
x = GlobalAveragePooling2D()(base_resnet.output)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)
model = Model(inputs=base_resnet.input, outputs=output)

model.compile(
    optimizer=Adam(1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()]
)

# Callbacks for early stopping and logging
callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ModelCheckpoint('best_resnet50_cpu.weights.h5', save_best_only=True, save_weights_only=True),
    CSVLogger('resnet50_cpu_training_log.csv')
]

# Train the model (no workers/use_multiprocessing arguments)
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=epochs,
    callbacks=callbacks
)
```

```python
import os
import cv2
import torch
from RRDBNet_arch import RRDBNet  # from ESRGAN repo folder
import numpy as np

# Set paths
esrgan_path = r'C:\Users\Apoorva\ESRGAN'  # Change if needed
model_path = os.path.join(esrgan_path, 'models', 'RRDB_ESRGAN_x4.pth')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Initialize model
model = RRDBNet(in_nc=3, out_nc=3, nf=64, nb=23, gc=32)
model.load_state_dict(torch.load(model_path), strict=True)
model.eval()
model = model.to(device)

target_size = (224, 224)  # desired image size
input_path = r'D:\Complete_Data\research\PCOS'
output_path_esrgan = r'D:\Complete_Data\research\PCOS_processed\esrgan'

def preprocess_esrgan_images(input_dir, output_dir):
    for label in ['infected', 'noninfected']:
        folder_in = os.path.join(input_dir, label)
        folder_out = os.path.join(output_dir, label)
        os.makedirs(folder_out, exist_ok=True)

        for file_name in os.listdir(folder_in):
            save_path = os.path.join(folder_out, file_name)
            
            # Skip already processed images
            if os.path.exists(save_path):
                print(f"Skipping {file_name} (already processed).")
                continue
            
            img_path = os.path.join(folder_in, file_name)
            img = cv2.imread(img_path)  # Read color image

            if img is None:
                print(f"Warning: Could not open {img_path}. Skipping.")
                continue

            img_resized = cv2.resize(img, target_size)
            img_rgb = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)
            img_tensor = torch.from_numpy(img_rgb).float().permute(2, 0, 1).unsqueeze(0).to(device)
            img_tensor /= 255.0  # Normalize

            with torch.no_grad():
                output = model(img_tensor).clamp(0, 1)

            output_img = output.squeeze(0).permute(1, 2, 0).cpu().numpy()
            output_img = (output_img * 255.0).round().astype(np.uint8)
            output_bgr = cv2.cvtColor(output_img, cv2.COLOR_RGB2BGR)

            cv2.imwrite(save_path, output_bgr)
            print(f"Processed and saved {file_name}")

preprocess_esrgan_images(input_path, output_path_esrgan)
```
